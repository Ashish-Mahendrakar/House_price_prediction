Predict the market price of a house automatically using linear regression model

```python
import pandas as pd  
import seaborn as sns
```

```python
data=pd.read_csv("HousePricePrediction.csv")
data.head()
```

```python
data["MSZoning"].unique()
```

```python
data["LotConfig"].unique()
```

```python
data["BldgType"].unique()
```

```python
data["Exterior1st"].unique()
```

```python
data=pd.get_dummies(data,columns=["MSZoning","LotConfig","BldgType","Exterior1st"],drop_first=True,dtype=int)
data.head()
```

```python
data.info()
```

```python
# Id column is not usefull

x=data.drop(columns=["SalePrice","Id"]) 

y=data["SalePrice"]
y.info()
```

```python
x.info()
```

```python
#split training and testing data
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
#linear regreesion model training
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
y_pred_train = reg.predict(x_train)
```

```python
r2_score(y_train, y_pred_train)
```
